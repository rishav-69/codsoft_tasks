```python
import tkinter as tk
import tkinter.simpledialog

class TodoApp:
    def __init__(self, root):
        self.root = root
        self.root.title("To-Do List")
        self.root.geometry("500x400")
        
        self.tasks = []
        
        self.title_label = tk.Label(root, text="To-Do List", font=("Times", 24, "bold italic"), fg="black", bg="green")
        self.title_label.pack(fill=tk.X, padx=10, pady=10)
        
        self.add_title_label = tk.Label(root, text="Add Items", font=("Helvetica", 12))
        self.add_title_label.pack(padx=10, anchor="w")
        
        self.add_frame = tk.Frame(root)
        self.add_frame.pack(padx=10, fill=tk.BOTH)
        
        self.add_entry = tk.Entry(self.add_frame, font=("Helvetica", 12))
        self.add_entry.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
        
        self.add_button = tk.Button(self.add_frame, text="Add", font=("Helvetica", 12), bg="blue", fg="white", command=self.add_task)
        self.add_button.pack(side=tk.LEFT)
        
        self.tasks_title_label = tk.Label(root, text="Tasks", font=("Helvetica", 12))
        self.tasks_title_label.pack(padx=10, pady=(20, 0), anchor="w")
        
        self.tasks_frame = tk.Frame(root)
        self.tasks_frame.pack(fill=tk.BOTH, padx=10, pady=10, expand=True)
        
        self.display_tasks()
    
    def add_task(self):
        task_text = self.add_entry.get()
        if task_text:
            self.tasks.append(task_text)
            self.add_entry.delete(0, tk.END)
            self.display_tasks()
    
    def edit_task(self, index):
        new_text = tk.simpledialog.askstring("Edit Task", "Edit the task:", initialvalue=self.tasks[index])
        if new_text:
            self.tasks[index] = new_text
            self.display_tasks()
    
    def delete_task(self, index):
        del self.tasks[index]
        self.display_tasks()
    
    def display_tasks(self):
        for widget in self.tasks_frame.winfo_children():
            widget.destroy()
        
        for index, task in enumerate(self.tasks):
            task_frame = tk.Frame(self.tasks_frame, bd=2, relief="groove")
            task_frame.grid(row=index, column=0, pady=(5, 0), sticky="we")
            
            task_label = tk.Label(task_frame, text=task, font=("Helvetica", 12), anchor="w")
            task_label.pack(side=tk.LEFT, padx=5)
            
            edit_button = tk.Button(task_frame, text="Edit", font=("Helvetica", 12),bg='green',fg='white', command=lambda i=index: self.edit_task(i))
            edit_button.pack(side=tk.LEFT, padx=10)
            
            delete_button = tk.Button(task_frame, text="Delete", font=("Helvetica", 12),bg='red',fg='white', command=lambda i=index: self.delete_task(i))
            delete_button.pack(side=tk.LEFT, padx=10)
    
if __name__ == "__main__":
    root = tk.Tk()
    app = TodoApp(root)
    root.mainloop()
```
