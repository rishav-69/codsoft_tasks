```python
import tkinter as tk
import string
import random

def generate_password():
    password_length = int(length_entry.get())
    
    if password_length <= 0:
        password_result.config(text="Invalid length")
        return
    
    characters = string.ascii_letters + string.digits + string.punctuation
    generated_password = ''.join(random.choice(characters) for _ in range(password_length))
    
    password_result.config(text=generated_password)

def accept_password():
    name = name_entry.get()
    password = password_result.cget("text")
    
    if name and password:
        with open("passwords.txt", "a") as file:
            file.write(f"Name: {name}\tPassword: {password}\n")
        
        name_entry.delete(0, tk.END)
        password_result.config(text="")
        status_label.config(text="Password saved successfully!", fg="green")
    else:
        status_label.config(text="Please generate a password first.", fg="red")

def reset_password():
    name_entry.delete(0, tk.END)
    length_entry.delete(0, tk.END)
    password_result.config(text="")
    status_label.config(text="")

root = tk.Tk()
root.geometry("400x350")
root.title("Password Generator")
root.configure(bg="white")

title_label = tk.Label(root, text="Password Generator", font=("Helvetica", 16, "bold"), fg="blue", bg="white")
title_label.pack(pady=10)

name_label = tk.Label(root, text="Enter User Name:", font=("Helvetica", 12), bg="white")
name_label.pack()

name_entry = tk.Entry(root, font=("Helvetica", 12))
name_entry.pack()

length_label = tk.Label(root, text="Enter Password Length:", font=("Helvetica", 12), bg="white")
length_label.pack()

length_entry = tk.Entry(root, font=("Helvetica", 12))
length_entry.pack()

generate_button = tk.Button(root, text="Generate Password", font=("Helvetica", 12), bg="blue", fg="white", command=generate_password)
generate_button.pack(pady=10)

password_result = tk.Label(root, text="", font=("Helvetica", 12), bg="white")
password_result.pack()

accept_button = tk.Button(root, text="Accept", font=("Helvetica", 12), bg="white", fg="blue", command=accept_password)
accept_button.pack(pady=5)

status_label = tk.Label(root, text="", font=("Helvetica", 12), bg="white")
status_label.pack()

reset_button = tk.Button(root, text="Reset", font=("Helvetica", 12), bg="white", fg="blue", command=reset_password)
reset_button.pack(pady=10)

root.mainloop()
```
